```python
import pandas as pd
import scipy.stats as stats
data = pd.read_csv('heart.csv')

column_stats = pd.DataFrame({
    "Mean": data.mean(),
    "Median": data.median(),
    "Std": data.std()
})

print("每項的統計數據（均值、中位數、標準差）：")
print(column_stats)

health_column = "target"
numerical_features = data.select_dtypes(include=['number']).columns.drop(health_column)

healthy = data[data[health_column] == 0]  
diseased = data[data[health_column] == 1] 
results = []
for feature in numerical_features:
    t_stat, p_value = stats.ttest_ind(healthy[feature], diseased[feature], equal_var=False)
    results.append({
        "Feature": feature,
        "T-Statistic": t_stat,
        "P-Value": p_value
    })
    
results_data = pd.DataFrame(results)
results_data["Significant"] = results_data["P-Value"] < 0.05

print("健康與患病群體的數值特徵比較結果：")
print(results_data)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler

data = pd.read_csv('heart.csv')

X = data.drop(columns=['target'])
y = data['target']

pearson_corr = X.corrwith(y)

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)  
chi2_selector = SelectKBest(score_func=chi2, k='all')
chi2_scores = chi2_selector.fit(X_scaled, y)

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X, y)
feature_importances = rf_model.feature_importances_

results = pd.DataFrame({
    'Feature': X.columns,
    'Pearson_Corr': pearson_corr.values,
    'Chi2_Score': chi2_scores.scores_,
    'RF_Importance': feature_importances
})

results.sort_values(by='RF_Importance', ascending=False, inplace=True)
print("特徵選擇結果：\n", results)

selected_features = results['Feature'].head(5).tolist()
print("\n選擇的特徵：", selected_features)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_curve, auc, RocCurveDisplay,classification_report,roc_auc_score
)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns



X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')

data = pd.read_csv('heart.csv')

if y_train.shape[1] == 1:
    y_train = y_train.iloc[:, 0]
if y_test.shape[1] == 1:
    y_test = y_test.iloc[:, 0]

rf_model = RandomForestClassifier(random_state=42)
lr_model = LogisticRegression(max_iter=1000, random_state=42)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {"Model": [], "Accuracy": [], "Precision": [], "Recall": [], "F1 Score": [], "AUC": []}

def evaluate_model(model, X, y, model_name):
    acc = cross_val_score(model, X, y, cv=cv, scoring='accuracy').mean()
    prec = cross_val_score(model, X, y, cv=cv, scoring='precision').mean()
    rec = cross_val_score(model, X, y, cv=cv, scoring='recall').mean()
    f1 = cross_val_score(model, X, y, cv=cv, scoring='f1').mean()
    auc_score = cross_val_score(model, X, y, cv=cv, scoring='roc_auc').mean()
    
    results["Model"].append(model_name)
    results["Accuracy"].append(acc)
    results["Precision"].append(prec)
    results["Recall"].append(rec)
    results["F1 Score"].append(f1)
    results["AUC"].append(auc_score)
    
evaluate_model(rf_model, X_train, y_train, "Random Forest")
evaluate_model(lr_model, X_train, y_train, "Logistic Regression")

result_df = pd.DataFrame(results)
print(result_df)

rf_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)

rf_probs = rf_model.predict_proba(X_test)[:, 1]
lr_probs = lr_model.predict_proba(X_test)[:, 1]

rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)

rf_auc = auc(rf_fpr, rf_tpr)
lr_auc = auc(lr_fpr, lr_tpr)

rcParams['font.sans-serif'] = ['DFKai-SB']  
rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10, 6))
plt.plot(rf_fpr, rf_tpr, label=f"Random Forest (AUC = {rf_auc:.2f})", linestyle="--")
plt.plot(lr_fpr, lr_tpr, label=f"Logistic Regression (AUC = {lr_auc:.2f})", linestyle="-")
plt.plot([0, 1], [0, 1], color="gray", linestyle=":")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC 曲線")
plt.legend(loc="lower right")
plt.grid(alpha=0.5)
plt.show()


y_pred_rf = rf_model.predict(X_test)
y_pred_lr = lr_model.predict(X_test)

y_prob_rf = rf_model.predict_proba(X_test)[:, 1]
y_prob_lr = lr_model.predict_proba(X_test)[:, 1]

def evaluate_on_test_set(y_test, y_pred, y_prob, model_name):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    
    print(f"模型: {model_name}")
    print(f"準確率 (Accuracy): {accuracy:.2f}")
    print(f"精確率 (Precision): {precision:.2f}")
    print(f"召回率 (Recall): {recall:.2f}")
    print(f"F1 分數: {f1:.2f}")
    print(f"AUC: {auc:.2f}")
    print("分類報告 (Classification Report):")
    print(classification_report(y_test, y_pred))
    print("\n")

evaluate_on_test_set(y_test, y_pred_rf, y_prob_rf, "Random Forest")
evaluate_on_test_set(y_test, y_pred_lr, y_prob_lr, "Logistic Regression")

cm_rf = confusion_matrix(y_test, y_pred_rf)
cmd_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=np.unique(y_test))
cmd_rf.plot(cmap='Blues')
plt.title('混淆矩陣 - Random Forest')
plt.show()

cm_lr = confusion_matrix(y_test, y_pred_lr)
cmd_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=np.unique(y_test))
cmd_lr.plot(cmap='Oranges')
plt.title('混淆矩陣 - Logistic Regression')
plt.show()

feature_importances = rf_model.feature_importances_
features = X_train.columns

plt.figure(figsize=(10, 6))
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.gca().invert_yaxis()
plt.xlabel('重要性')
plt.title('特徵重要性 (Random Forest)')
plt.show()

```

```python
# 1. 載入資料
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')

# 將目標值轉換為 Series
if y_train.shape[1] == 1:
    y_train = y_train.iloc[:, 0]
if y_test.shape[1] == 1:
    y_test = y_test.iloc[:, 0]

# 定義模型
lr_model = LogisticRegression(max_iter=1000, random_state=42)
rf_model = RandomForestClassifier(random_state=42)

# 訓練模型
lr_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)

# 預測測試集
y_pred_lr = lr_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)

# 添加 `actual` 和預測值到測試集
X_test['actual'] = y_test.values
X_test['predicted_lr'] = y_pred_lr
X_test['predicted_rf'] = y_pred_rf

# 提取假陽性和假陰性樣本
false_positives_lr = X_test[(X_test['actual'] == 0) & (X_test['predicted_lr'] == 1)]
false_negatives_lr = X_test[(X_test['actual'] == 1) & (X_test['predicted_lr'] == 0)]

false_positives_rf = X_test[(X_test['actual'] == 0) & (X_test['predicted_rf'] == 1)]
false_negatives_rf = X_test[(X_test['actual'] == 1) & (X_test['predicted_rf'] == 0)]

# 設置中文字體
rcParams['font.sans-serif'] = ['DFKai-SB']
rcParams['axes.unicode_minus'] = False

# 取得所有數值型特徵
numeric_features = X_test.select_dtypes(include=[np.number]).columns.tolist()
numeric_features = [col for col in numeric_features if col not in ['actual', 'predicted_lr', 'predicted_rf']]

# 統計計算函數
def calculate_statistics(feature, data, false_positives, false_negatives):
    if feature not in data.columns:
        print(f"特徵 {feature} 不存在於數據框中，跳過。")
        return
    print(f"\n特徵: {feature}")
    print(f"整體數據 - 均值: {data[feature].mean():.2f}, 中位數: {data[feature].median():.2f}, 標準差: {data[feature].std():.2f}")
    if not false_positives.empty:
        print(f"假陽性 - 均值: {false_positives[feature].mean():.2f}, 中位數: {false_positives[feature].median():.2f}, 標準差: {false_positives[feature].std():.2f}")
    else:
        print(f"假陽性 - 無數據")
    if not false_negatives.empty:
        print(f"假陰性 - 均值: {false_negatives[feature].mean():.2f}, 中位數: {false_negatives[feature].median():.2f}, 標準差: {false_negatives[feature].std():.2f}")
    else:
        print(f"假陰性 - 無數據")

# 繪圖函數
def plot_feature_distribution(feature, data, false_positives, false_negatives, model_name):
    if feature not in data.columns:
        print(f"特徵 {feature} 不存在於數據框中，跳過繪圖。")
        return

    # KDE 分佈圖
    plt.figure(figsize=(10, 5))
    sns.kdeplot(data[feature], label="整體數據", color="blue", fill=True)
    if not false_positives.empty:
        sns.kdeplot(false_positives[feature], label="假陽性", color="red", fill=True)
    if not false_negatives.empty:
        sns.kdeplot(false_negatives[feature], label="假陰性", color="green", fill=True)
    plt.title(f"{feature} 分佈圖 - {model_name}")
    plt.xlabel(feature)
    plt.ylabel("密度")
    plt.legend()
    plt.show()

    

# 對所有數值型特徵進行分析
for feature in numeric_features:
    print(f"處理特徵: {feature}")
    # Logistic Regression
    print("=== Logistic Regression ===")
    calculate_statistics(feature, X_test, false_positives_lr, false_negatives_lr)
    plot_feature_distribution(feature, X_test, false_positives_lr, false_negatives_lr, "Logistic Regression")

    # Random Forest
    print("=== Random Forest ===")
    calculate_statistics(feature, X_test, false_positives_rf, false_negatives_rf)
    plot_feature_distribution(feature, X_test, false_positives_rf, false_negatives_rf, "Random Forest")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')

data = pd.read_csv('heart.csv')

rf_model = RandomForestClassifier(random_state=42)
lr_model = LogisticRegression(max_iter=1000, random_state=42)

train_sizes, train_scores_rf, test_scores_rf = learning_curve(
    rf_model, X_train, y_train, cv=5, train_sizes=np.linspace(0.1, 1.0, 10), scoring='accuracy'
)

train_sizes, train_scores_lr, test_scores_lr = learning_curve(
    lr_model, X_train, y_train, cv=5, train_sizes=np.linspace(0.1, 1.0, 10), scoring='accuracy'
)

train_mean_rf = np.mean(train_scores_rf, axis=1)
train_std_rf = np.std(train_scores_rf, axis=1)
test_mean_rf = np.mean(test_scores_rf, axis=1)
test_std_rf = np.std(test_scores_rf, axis=1)

train_mean_lr = np.mean(train_scores_lr, axis=1)
train_std_lr = np.std(train_scores_lr, axis=1)
test_mean_lr = np.mean(test_scores_lr, axis=1)
test_std_lr = np.std(test_scores_lr, axis=1)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 2)
plt.plot(train_sizes, train_mean_rf, label='Training score', color='blue')
plt.plot(train_sizes, test_mean_rf, label='Validation score', color='orange')
plt.fill_between(train_sizes, train_mean_rf - train_std_rf, train_mean_rf + train_std_rf, color='blue', alpha=0.2)
plt.fill_between(train_sizes, test_mean_rf - test_std_rf, test_mean_rf + test_std_rf, color='orange', alpha=0.2)
plt.title('Learning Curve: Random Forest')
plt.xlabel('Training Size')
plt.ylabel('Accuracy')
plt.legend(loc='best')

plt.subplot(1, 2, 1)
plt.plot(train_sizes, train_mean_lr, label='Training score', color='blue')
plt.plot(train_sizes, test_mean_lr, label='Validation score', color='orange')
plt.fill_between(train_sizes, train_mean_lr - train_std_lr, train_mean_lr + train_std_lr, color='blue', alpha=0.2)
plt.fill_between(train_sizes, test_mean_lr - test_std_lr, test_mean_lr + test_std_lr, color='orange', alpha=0.2)
plt.title('Learning Curve: Logistic Regression')
plt.xlabel('Training Size')
plt.ylabel('Accuracy')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```
